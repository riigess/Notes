# Generate Paranthesis

You are given an integer `n`. Return all well-formed parantheses strings you can generate with `n` pairs of parenthesis.

## Example 1
```python
Input: n = 1
Output: ["()"]
```

## Example 2
```python
Input: n = 3
Output: ["((()))", "(()())", "(())()", "()(())", "()()()"]
```

You may return the answer in **any** order.

### Constraints
- `1 <= n <= 7`

````python
from typing import List


class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        stack = []
        ret = []
        def backtrack(o, c):
            if o == c == n:
                ret.append("".join(stack))
                return
            if o < n:
                stack.append("(")
                backtrack(o + 1, c)
                stack.pop()
            if c < o:
                stack.append(")")
                backtrack(o, c + 1)
                stack.pop()
        backtrack(0, 0)
        return ret
````

````python
#Testing
sol = Solution()
assert sol.generateParenthesis(1) == ["()"]
assert sol.generateParenthesis(3) == ["((()))","(()())","(())()","()(())","()()()"]
print("All tests passed")
````
